```python
import pandas as pd
```

```python
news = 'Data/news.csv'
pdf = 'Data/pdf.csv'
web = 'Data/web.csv'

web_data = pd.read_csv(web)
news_data = pd.read_csv(news)
pdf_data = pd.read_csv(pdf)

web_data
```

```python
web_data = pd.read_csv(web)
news_data = pd.read_csv(news)
pdf_data = pd.read_csv(pdf)

for i in web_data, news_data, pdf_data :
    i.drop(i.columns[0], axis=1, inplace=True)

pdf_data.drop(columns='URLs', inplace=True)

pdf_data = pdf_data.rename(columns={"Title": "instruction", "Data": "output"})

news_data = news_data.rename(columns={"Title_Date": "instruction", "DATA": "output"})

news_data
df = pd.DataFrame(columns=['instruction', 'input', 'output'])

df = pd.concat([df, pdf_data])
df = pd.concat([df, news_data])

file_path = 'data_news.json'

df.to_json(file_path, orient='records')

df.fillna('', inplace=True)

df
```
